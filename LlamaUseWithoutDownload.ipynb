```python
from google.colab import userdata
hf_token = userdata.get('HF_token')
```

```python
import time
from huggingface_hub import InferenceClient


# 1. Initialize the client
# Replace 'YOUR_HF_TOKEN' with the token from https://huggingface.co/settings/tokens
client = InferenceClient(
    model="meta-llama/Llama-3.2-1B-Instruct",
    api_key=hf_token
)

def generate_text(user_input):
    system_prompt = (
        "You are a viral content expert. Rewrite the provided text to be high-engagement. "
        "Return ONLY the rewritten text. No greetings, no explanations, no preamble, and no conversational filler."
    )
    try:
        results = []
        for i in range(5):
            output = client.chat.completions.create(
                model="meta-llama/Llama-3.2-1B-Instruct",
                messages=[
                    {"role": "system", "content": system_prompt},
                    {"role": "user", "content": f"Rewrite this: {user_input}"}
                ],
                temperature=0.9,
                max_tokens=400
            )

            clean_response = output.choices[0].message.content.strip()
            results.append(clean_response)

            time.sleep(1)

        return results

    except Exception as e:
        if "503" in str(e):
            print("Model is loading on server... waiting 10 seconds.")
            time.sleep(10)
            return generate_text(user_input)
        else:
            return f"An error occurred: {e}"
```

```python
user_input = "AI is making agriculture better"
```

```python
output = generate_text(user_input)

print(output)
```
