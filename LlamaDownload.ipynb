```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "meta-llama/Llama-3.2-1B"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)
```

```python
model.save_pretrained("llama_model")
tokenizer.save_pretrained("llama_model")
```

<h2> Using model to generate text </h2>

```python
prompt = "Rewrite this text to make it more engaging and viral: AI helps farmers improve crop yield."

inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

output = model.generate(
    **inputs,
    max_new_tokens=60,
    temperature=0.9,
    num_return_sequences=5
)

for i, o in enumerate(output):
    print(f"Version {i+1}:")
    print(tokenizer.decode(o, skip_special_tokens=True))
    print()
```
